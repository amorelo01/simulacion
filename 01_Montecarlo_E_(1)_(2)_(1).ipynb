# **¬øQue es simulaci√≥n?**

##**La simulaci√≥n abstrae el comportamiento del mundo real**


*   Modelo es la representaci√≥n de un sistema mediante un conjunto relaciones l√≥gico-matem√°ticas originadas en la observaci√≥n o percepci√≥n del comportamiento del sistema.
*   La simulaci√≥n de un sistema es la creaci√≥n de un modelo que abstrae el comportamiento del mundo real y que puede ser usado para generar comportamientos artificiales del sistema, de forma tal que permite replicar y predecir diversas situaciones del comportamiento del mismo.
*   Los comportamientos artificiales son experimentos.

##**$X$ simula a $Y$**

*   La propuesta de definici√≥n de simulaci√≥n propuesta por Charles  West Churchman es ajustadamente formal*:
"X simula a Y" s√≠ y solo s√≠:

>1.   $X$ e $Y$ son sistemas formales.
>2.   $Y$ se considera como sistema real
>3.   $X$ se toma como una aproximaci√≥n del sistema real
>4.   Las reglas de validez en $X$ no est√°n exentas de error.

<strong>*</strong> *Un sistema formal o sistema l√≥gico es un sistema abstracto compuesto por un lenguaje formal, axiomas, reglas de inferencia y a veces una sem√°ntica formal, que se utiliza para deducir o demostrar teoremas y dar una definici√≥n rigurosa del concepto de demostraci√≥n.*

Simulaci√≥n de un sistema (o un organismo) es la operaci√≥n de un modelo (simulador), el cual es una representaci√≥n del sistema. Este modelo puede sujetarse a manipulaciones que ser√≠an imposibles de realizar, demasiado costosas o impr√°cticas.

# **Importancia de la simulaci√≥n**

##**La simulaci√≥n es una de las t√©cnicas mas usadas**

Recientes avances en las metodolog√≠as de simulaci√≥n y la gran disponibilidad de software que actualmente existe en el mercado, han hecho que la t√©cnica de simulaci√≥n sea una de las herramientas m√°s ampliamente usadas en el an√°lisis de sistemas. Thomas H. Naylor ha sugerido que un estudio de simulaci√≥n es muy importante para la ingenier√≠a porque presenta las siguientes ventajas en el dise√±o de estos(B√∫, 1994).

##**La simulaci√≥n permite un mejor entendimiento del sistema**

*   A trav√©s de un estudio de simulaci√≥n, se puede estudiar el efecto de cambios internos y externos del sistema, al hacer alteraciones en el modelo del sistema y observando los efectos de esas alteraciones en el comportamiento del sistema.
*   Una observaci√≥n detallada del sistema que se est√° simulando puede conducir a un mejor entendimiento del sistema y por consiguiente a sugerir estrategias que mejoren la operaci√≥n y eficiencia del sistema(B√∫, 1994).

##**La simulaci√≥n permite ver nuevas situaciones**

*   La simulaci√≥n de sistemas complejos puede ayudar a entender mejor la operaci√≥n del sistema, a detectar las variables m√°s importantes que interact√∫an en el sistema y a entender mejor las interrelaciones entre estas variables.
*   La t√©cnica de simulaci√≥n puede ser utilizada para experimentar con nuevas situaciones, sobre las cuales se tiene poca o ninguna informaci√≥n. A trav√©s de esta experimentaci√≥n se puede anticipar mejoras a posibles resultados no previstos.
*   Cuando nuevos elementos son introducidos en un sistema, la simulaci√≥n puede ser usada para anticipar cuellos de botella o alg√∫n otro problema que puede surgir en el comportamiento del sistema.(B√∫, 1994).

# **Aplicaciones de la simulaci√≥n**

##**La simulaci√≥n es conveniente cuando‚Ä¶**

Para comenzar, se puede analizar una serie de situaciones en las cuales es apropiado el uso de la simulaci√≥n. Estas  situaciones fueron puestas de manifiesto por Shannon (1975)*. La simulaci√≥n es conveniente cuando:
*   No existe una formulaci√≥n matem√°tica anal√≠ticamente resoluble. Muchos sistemas reales no pueden ser modelados matem√°ticamente con las herramientas actualmente disponibles; por ejemplo, la conducta de un cliente de un banco.
*   Existe una formulaci√≥n matem√°tica, pero es dif√≠cil obtener una soluci√≥n anal√≠tica. Los modelos matem√°ticos utilizados para modelar un reactor nuclear o una planta qu√≠mica son imposibles de resolver en forma anal√≠tica sin realizar serias simplificaciones.
*   No existe el sistema real. Es problema del ingeniero, que tiene que dise√±ar un sistema nuevo; el dise√±o del sistema mejorar√° notablemente si se cuenta con un modelo adecuado para realizar experimentos.
*   Los experimentos son imposibles debido a impedimentos econ√≥micos, de seguridad, de calidad o √©ticos. En este caso, el sistema real est√° disponible para realizar experimentos, pero la dificultad de estos hace que se descarte esta opci√≥n. Un ejemplo de esto es la imposibilidad de provocar fallas en un avi√≥n real para evaluar la conducta del piloto.
*   El sistema evoluciona muy lentamente o muy r√°pidamente. Un ejemplo de din√°mica lenta es el problema de los cient√≠ficos que estudian la evoluci√≥n del clima. Ellos deben predecir la conducta futura del clima, pero, dadas las condiciones actuales, no pueden esperar a que un tornado arrase una ciudad para luego dar el mensaje de alerta. Por el contrario, existen fen√≥menos muy r√°pidos que deben ser simulados para poder observarse en detalle, por ejemplo, una explosi√≥n.

##**Pasos de un proyecto de simulaci√≥n**

El proceso para creaci√≥n y desarrollo de un modelo de simulaci√≥n se pude resumir en los siguientes pasos:
1.   **Formulaci√≥n del problema**.  En este paso se debe comenzar con una descripci√≥n del problema o del sistema. Debe existir una correcta identificaci√≥n del objetivo, las variables de decisi√≥n, las restricciones, la medida de efectividad y las variables no controlables y su comportamiento estad√≠stico. En este paso debe quedar perfectamente establecido el objeto de la simulaci√≥n.
2.   **Recolecci√≥n de datos**. Se deben describir las interacciones entre las variables de decisi√≥n, de tal suerte que interact√∫en dentro del sistema; lo mismo con las restricciones del sistema y las medidas de efectividad que se van utilizar para definir y estudiar el sistema y los resultados que se espera obtener. La naturaleza y cantidad de datos necesarios est√°n determinadas por la formulaci√≥n del problema y del modelo. Los datos pueden ser provistos por:
>*   Registros hist√≥ricos.
>*   Experimentos de laboratorios.
>*   Mediciones realizadas en el sistema real.

3.   **Construcci√≥n del modelo**. En este paso se genera un modelo de simulaci√≥n de base con el cual se obtendr√°n los  resultados deseados mediante las interacciones entre las variables que forman parte del sistema y sus relaciones  l√≥gico-matem√°ticas que describan en forma completa al modelo.
>*   Verificaci√≥n:
>>*   Consiste en analizar el modelo con entidades familiarizadas con la operaci√≥n del sistema.
>>*   Es el proceso de demostraci√≥n de que el modelo trabaja como se intent√≥.
>>*   Lenguaje por utilizar o qu√© paquete, para procesarlo en la computadora y obtener los resultados deseados.

4.   **Realizaci√≥n de corridas piloto**. Obtener el tama√±o de la muestra de simulaci√≥n que depende generalmente del tipo de distribuci√≥n que se intenta simular. Este paso se divide en:
>*   Codificaci√≥n del modelo: consiste en generar las instrucciones o c√≥digo computacional necesario para lograr que el modelo pueda ser ejecutado en un ambiente computacional.
>*   Validaci√≥n del modelo: a trav√©s de este paso es posible detallar deficiencias en la formulaci√≥n del modelo o en los datos alimentados a este. La validaci√≥n se lleva a cabo mediante la comparaci√≥n estad√≠stica entre los resultados del modelo y los resultados reales.

5.   **Dise√±o de experimentos**. La experimentaci√≥n determina las diversas alternativas que deben ser evaluadas,  seleccionando las variables de entrada y sus diferentes niveles, con la finalidad de optimizar las variables de  respuestas del sistema real. En este paso se utilizan las t√©cnicas de dise√±o de experimento y superficies de respuestas para realizar an√°lisis de sensibilidad de los √≠ndices requeridos.

6.   **An√°lisis de datos de salida**. Una vez seleccionada la mejor alternativa, es importante generar el modelo final para llevarlo a la pr√°ctica, listo para realizar la simulaci√≥n y estudiar el comportamiento del proceso

## La regla de los ‚Äútres sigmas‚Äù

La regla de los "tres sigmas", tambi√©n conocida como la regla 68-95-99.7, es un principio estad√≠stico que se aplica a distribuciones de datos que siguen una curva normal (distribuci√≥n normal o gaussiana). Esta regla establece lo siguiente en relaci√≥n con la desviaci√≥n est√°ndar (œÉ) de la media (Œº) de la distribuci√≥n:
*   Alrededor del 68% de los datos se encuentra dentro de una desviaci√≥n est√°ndar de la media. Es decir, entre Œº - œÉ y Œº + œÉ.
*   Alrededor del 95% de los datos se encuentra dentro de dos desviaciones est√°ndar de la media. Es decir, entre Œº - 2œÉ y Œº + 2œÉ.
*   Alrededor del 99.7% de los datos se encuentra dentro de tres desviaciones est√°ndar de la media. Es decir, entre Œº - 3œÉ y Œº + 3œÉ.

Esta regla es √∫til para comprender c√≥mo se distribuyen los valores en una distribuci√≥n normal, permitiendo estimar la variabilidad de los datos y la probabilidad de que ciertos valores caigan dentro de un rango espec√≠fico alrededor de la media. Se utiliza ampliamente en varios campos, incluidos los negocios, la ingenier√≠a, las ciencias naturales y sociales, entre otros, para analizar y hacer predicciones basadas en patrones de datos.

# **Montecarlo Integrales**
## Estimar integrales usando Montecarlo

La simulaci√≥n mediante m√©todos MonteCarlo ha sido muy utilizada en el c√°lculo de integrales, supongamos $f(x)$ una funci√≥n en la cual deseamos calcular su integral definida entre $0$ y $1$.

$\theta = \int_{a}^{b} \ f(x) \ dx$


Para calcular el valor de $\theta$, tenemos a $U$ distribuida uniformemente entre $(0,1)$, entonces podemos expresar a $\theta$ como:

$\theta = E[f(U)]$

Si $U_1$, $U_2$, ..., $U_k$ son variables aleatorias independientes y uniformes en $(0,1)$, esto implica que $f(U_1)$, $f(U_2)$, ..., $f(U_k)$ son variables aleatorias independientes e id√©nticamente distribuidas con media $\theta$. Por la ley fuerte de los grandes n√∫meros, se tiene, con probabilidad 1,

$\sum_{i=1}^{n} \frac{f(U_1)}{k}\rightarrow E[f(U)]=\theta \ \ cuando \ k\rightarrow \infty$


Por lo tanto se puede estimar $\theta$ generando un gran n√∫mero de n√∫meros aleatorios $u_i$ y considerando nuestra aproximaci√≥n a $\theta$ el valor promedio de $f(u)$


## Ley de los grandes n√∫meros

*   En la teor√≠a de la probabilidad, bajo el t√©rmino gen√©rico de ley de los grandes n√∫meros se engloban varios teoremas que describen el comportamiento del promedio de una sucesi√≥n de variables aleatorias conforme aumenta su n√∫mero de ensayos.
*   Estos teoremas prescriben condiciones suficientes para garantizar que dicho promedio converja al promedio de las esperanzas de las variables aleatorias involucradas. Las distintas formulaciones de la ley de los grandes n√∫meros (y sus condiciones asociadas) especifican la convergencia de formas distintas.
*   Las leyes de los grandes n√∫meros explican por qu√© el promedio de una muestra al azar de una poblaci√≥n de gran tama√±o tender√° a estar cerca de la media de la poblaci√≥n completa.
*   Cuando las variables aleatorias tienen una varianza finita, el teorema del l√≠mite central extiende nuestro entendimiento de la convergencia de su promedio describiendo la distribuci√≥n de diferencias estandarizadas entre la suma de variables aleatorias y el valor esperado de esta suma: sin importar la distribuci√≥n subyacente de las variables aleatorias, esta diferencia estandarizada converge a una variable aleatoria normal est√°ndar.
*   La frase "ley de los grandes n√∫meros" es tambi√©n usada ocasionalmente para referirse al principio de que la probabilidad de que cualquier evento posible (incluso uno improbable) ocurra al menos una vez en una serie, aumenta con el n√∫mero de eventos en la serie. Por ejemplo, la probabilidad de que un individuo gane la loter√≠a es bastante baja; sin embargo, la probabilidad de que alguien gane la loter√≠a es bastante alta, suponiendo que suficientes personas comprasen boletos de loter√≠a.

# **Caminata Aleatoria**

**Una caminata aleatoria es una ruta definida como una serie de pasos aleatorios**

Las caminatas aleatorias pueden utilizarse para modelar fen√≥menos que ocurren en el mundo real, desde movimientos de mol√©culas en un gas hasta el comportamiento de un apostador que pasa un d√≠a en el casino.
Una caminata aleatoria es una sucesi√≥n de variables aleatorias reales $(S_n)_{n=0}^\infty$ tal que $S_0 = 0$ y las variables aleatorias $(S_i - S_{i-1})_{n=0}^\infty$ son independientes e id√©nticamente distribuidas. Si $X_i = S_i-S_{i-1}$, entonces a la distribuci√≥n de $X_i$, que es independiente de $i$, y a la que daremos el nombre de distribuci√≥n del salto, gobierna a la caminata aleatoria y las caracter√≠sticas de esta √∫ltima se obtienen a trav√©s del estudio de dicha distribuci√≥n.

Si $X_i$ tiene una distribuci√≥n Bernoulli sim√©trica de par√°metro $1/2$ se le conoce como caminata aleatoria simple y ciertas distribuciones relacionadas con esta sucesi√≥n se encuentran mediante m√©todos combinatorios; sin embargo, para el caso general, los m√©todos a utilizar generalmente son m√°s complicados y el objetivo de su estudio es expresar la soluci√≥n al problema planteado en t√©rminos de las caminatas aleatorias a trav√©s de la funci√≥n de distribuci√≥n. A veces tal soluci√≥n es muy complicada y no es de utilidad pr√°ctica, por lo que adem√°s de la soluci√≥n exacta, generalmente buscamos expresiones aproximadas m√°s sencillas. Este enfoque se puede utilizar aun cuando la soluci√≥n exacta a nuestro problema nos sea desconocida.

Nota: Valor adicional, graficar los puntos en un plano que muestre las circunferencias de diferente color y los puntos por color.

El c√≥digo se muestra a continuaci√≥n:



```python
import random as rnd
TURNOS = 10
RONDAS = 20
APUESTA = 100
def puntos_dardo():
    x, y = rnd.uniform(-1, 1), rnd.uniform(-1, 1)
    distancia_al_centro = x**2 + y**2
    if distancia_al_centro <= 0.2**2:
        return 50
    elif distancia_al_centro <= 0.5**2:
        return 25
    elif distancia_al_centro <= 1:
        return 10
    else:
        return 0
def jugar_ronda():
    puntos_jugador1 = sum(puntos_dardo() for _ in range(TURNOS))
    puntos_jugador2 = sum(puntos_dardo() for _ in range(TURNOS))
    if puntos_jugador1 > puntos_jugador2:
        return APUESTA
    elif puntos_jugador2 > puntos_jugador1:
        return -APUESTA
    else:
        return 0
def juego_noche():
    ganancias_jugador1 = 0
    ganancias_jugador2 = 0
    for _ in range(RONDAS):
        ganancia_ronda = jugar_ronda()
        ganancias_jugador1 += ganancia_ronda
        ganancias_jugador2 -= ganancia_ronda  # Si el jugador 1 gana, el jugador 2 pierde, y viceversa
    return ganancias_jugador1, ganancias_jugador2
if __name__ == "__main__":
    ganancias_j1, ganancias_j2 = juego_noche()
    print(f"Ganancias Jugador 1: {ganancias_j1} unidades")
    print(f"Ganancias Jugador 2: {ganancias_j2} unidades")
```
